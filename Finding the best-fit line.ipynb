### Scenario

A venture capital firm has hired you to evaluate e-commerce investment opportunities. To inform an upcoming quarterly report, the firm needs to know the future median value of a business it sees as a potential future investment.

Note: A venture capital firm invests capital in businesses that have high growth potential, in exchange for owning a part of those businesses.

```python
# Install the statsmodels package. 
!pip install statsmodels
# Import necessary libraries, packages, and classes 
import pandas as pd
import numpy as np
# Visualisation
import matplotlib.pyplot as plt
import seaborn as sns 
# Linear Regression
import sklearn
import statsmodels.api as sm 
```

```python
# Import and read the data file
df = pd.read_csv("ecommerce_data.csv")
# Replace all missing values with 0
df. replace(np. nan,0)

# View the DataFrame 
print(df.head())
print(df.info())
df.describe()
```

```python
df.head()
```

## Prepare the Data

```python
# Define the dependent variable.
y = df['Median_s'].values.reshape(-1, 1) 

# Define the independent variable.
x = df['avg_no_it'].values.reshape(-1, 1) 
```

## Data Partitioning

```python
# Import the sklearn module.
from sklearn.model_selection import train_test_split 

# Create  the subset (70/30); 
# Control the shuffling/avoid variation in values between variable.
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size=0.3,
                                                    random_state=42)
```

## Fit the Model 

```python
# Import the necessary package.
from sklearn.linear_model import LinearRegression

# Specify the model.
lm = LinearRegression()

# Fit the model.
lm.fit(x_train, y_train)
```

## Predicting the test data

```python
# Predict the training set values.
y_pred = lm.predict(x_test) 

# View the output
y_pred
```

## Visualise the simple linear regression model

```python
# Visualise the training set.
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, lm.predict(x_train), color = 'green')
plt.title("Avg no of items vs Median of Seller Business(Training Data)")
plt.xlabel("Avg no of items")
plt.ylabel("Median of Seller Business")

plt.show()
```

```python
# Visualise the test set.
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, lm.predict(x_train), color = 'green')
plt.title("Avg no of items vs Median of Seller Business(Testing Data)")
plt.xlabel("Avg no of items")
plt.ylabel("Median of Seller Business")

plt.show()
```

## Accuracy of train data

```python
# Print R-squared value of the test data.
print("R-squared value: ", lm.score(x_train, y_train)) 

# Print the intercept value.
print("Intercept value: ", lm.intercept_) 

# Print the coefficient value.
print("Coefficient value: ", lm.coef_)
```

## Summary

> What does the summary indicates?
> - $R^2$: 49% of the total variability of `y` (median value of each business), is explained by the variability of `X` (items sold on average).  
> - The coefficient value of 9.12 tells us that as the Avg_no_it variable increases by 1, the predicted value of Median_s increases by 9.12.
